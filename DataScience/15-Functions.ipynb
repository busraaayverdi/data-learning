```python
def args_sum(*args):
    return sum(args)
```

```python
args_sum(10,20,30,40,50)
```

```python
def args_example(*args):
    print(args)
```

```python
args_example(123,120123123,353,2144,423)
```

```python
 def kwargs_example(**kwargs):
     print(kwargs)
```

```python
kwargs_example(apple =100, banana =150,melon = 200)
```

```python
def kwargs_example_2(**kwargs):
    if "apple" in kwargs:
        print("appleeeeeeeeeee")
    else:
        print(":(")
```

```python
kwargs_example_2(apple = 100,banana =150,melon = 200)
```

```python
kwargs_example_2(banana=150)
```
