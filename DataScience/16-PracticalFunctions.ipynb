```python
def divideNumber(number):
    return number / 2
```

```python
divideNumber(20)
```

```python
myList = [3,5,7,10,20,30]
```

```python
myResultList = [divideNumber(number) for number in myList]
print(myResultList)
```

```python
myResultList= []
for number in myList:
    myResultList.append(divideNumber(number))
print(myResultList)
```

```python
def controlString(string):
    return "Atil" in string
```

```python
controlString("Atil Samancioglu")
```

```python
controlString("busra")
```

```python
myStringList = ["Atil","Busra","Atil samancığlu"]
```

```python
list(map(controlString, myStringList))
```

FILTER

```python
#filter(controlString, myStringList)
list(filter(controlString, myStringList))
```

LEGB -> L->Local, E-> Enclosing, G->Global, B->Built-In

```python
#Global

myString = "Atıl"

def myFunction():
    #enclosing
    myString = "Atıl 2"
    print(myString)

    def myFunction2():
        #Local
        myString = "Atıl 3"
        print(myString)

    myFunction2()
```
