```python
import numpy as np
from pandas.core.reshape.concat import new_axes
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,60,70]
```

```python
weight_list= [20,60,80,85,70,90,95,120]
```

```python
plt.plot(age_list,weight_list,"g") #y -> yellow g,b,r
plt.xlabel("age")
plt.ylabel("weight")
plt.title("Age vs Weight")
plt.show()
# plt.savefig("test.png")
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2,"*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1,numpy_arr2,"*-")
plt.subplot(1,2,2)
plt.plot(numpy_arr2,numpy_arr1,"*-")
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.5,0.3])
figure_axes.plot(numpy_arr1,numpy_arr2,"g")
figure_axes.set_xlabel("age")
figure_axes.set_ylabel("weight")
figure_axes.set_title("Age vs Weight")
plt.show()
```

```python
new_fig = plt.figure(dpi = 100)
new_axes = new_fig.add_axes([0.1,0.1,0.5,0.3])
new_axes.plot(numpy_arr1,numpy_arr1 **2,label = "numpy array **2")
new_axes.plot(numpy_arr1,numpy_arr1 **3,label = "numpy array **3")
new_axes.legend()
plt.show()
```
