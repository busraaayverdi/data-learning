```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data1
```

```python
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
 my_fig , my_axes = plt.subplots()
```

```python
type(my_axes)
```

```python
type(my_fig)
```

```python
my_fig , my_axes = plt.subplots()
my_axes.plot(data1, data2, color = "#1E0703")
my_axes.plot(data2, data1, color = "#53CA0A", alpha = 0.3)
plt.show()
```

```python
(new_fig , new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2, color = "blue", linewidth = 2)
new_axes.plot(data1, data1 + 3, color = "yellow", linewidth = 5)
new_axes.plot(data1, data1 + 4, color = "pink", linewidth = 2 ,linestyle = "--")
new_axes.plot(data1, data1 + 5, color = "pink", linewidth = 2 ,linestyle = "-.")
new_axes.plot(data1, data1 + 6, color = "pink", linewidth = 2 ,linestyle = ":")
new_axes.plot(data1, data1 + 7, color = "black",linestyle = "--",marker = "o",markersize = 5 ,markerfacecolor = "red")
new_axes.plot(data1, data1 + 9, color = "black",linestyle = "--",marker = "o",markersize = 10)
plt.show()
```
